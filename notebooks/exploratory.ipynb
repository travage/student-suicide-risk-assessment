```python
# For data manipulation
import pandas as pd
import numpy as np

# For plotting
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots appear inside the notebook
%matplotlib inline

df = pd.read_csv("../data/raw/Final_SP_dataSet.csv", keep_default_na=False)
```

# Data Cleaning

## Step 1: Convert column names to lowercase and snake case

```python
# See basic info: columns, data types, and null counts
print(df.info())

# Convert column names to lowercase and snake case
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print(df.info())
```

## Step 2: Verify columns are correct data types

```python
for col in df.columns:
    print(f"Column: {col}")
    print(df[col].apply(type).value_counts())
    print("-" * 40)
```

## Step 3: Check unique values in columns for missing/incorrect data

```python
for col in df.columns:
    print(df[col].value_counts())
```

```python
# Age of 221 is incorrect
# Changing it to the median age
median_age = df.loc[df['age'] < 100, 'age'].median()
df.loc[df['age'] == 221, 'age'] = median_age

# Found invalid answer 'pf' in depression_level column
# Changing it to 'Missing'
df.loc[df['depression_level'] == 'pf', 'depression_level'] = 'Missing'

# Clean strings
for col in df.columns:
    if col == 'age':
        continue
    else:
        df[col] = df[col].str.strip().str.title()

for col in df.columns:
    print(df[col].value_counts())
```

## Step 4: Encode categorical variables

### Ordinal Categories

```python
# Ordinal columns
ordinal_mappings = {
    'stress_level': {
        'Low': 0,
        'Moderate': 1,
        'High': 2
    },
    'academic_performance': {
        'Excellent': 0,
        'Good': 1,
        'Average': 2,
        'Poor': 3
    },
    'health_condition': {
        'Normal': 0,
        'Fair': 1,
        'Abnormal': 2
    },
    'depression_level': {
        'Sometimes': 0,
        'Often': 1,
        'Always': 2
    },
    'anxiety_level': {
        'Sometimes': 0,
        'Often': 1,
        'Always': 2
    },
    'self_harm_story': {
        'No': 0,
        'Yes': 1
    },
    'suicide_attempt': {
        'Never Thought': 0,
        'Thought': 1,
        'Attempted': 2
    }
}

# Apply ordinal mappings to the dataframe
for column, mapping in ordinal_mappings.items():
    df[column] = df[column].map(mapping)

# Set missing value to -1 and convert to int
df['depression_level'] = df['depression_level'].fillna(-1).astype(int)

for col in df.columns:
    print(df[col].value_counts())
```

### Nominal Categories

```python
nominal_columns = ['gender', 'relationship_condition', 'family_problem', 'mental_support']
# One-hot encoding of nominal columns; preserving all dummy columns to preserve interpretability
df = pd.get_dummies(df, columns=nominal_columns, drop_first=False)
print(df.columns)
```

```python
for col in df.columns:
    print(df[col].value_counts())
```

```python
df.isnull().sum().sort_values(ascending=False)
```

# Model Training

```python
# Set features (x) and target (y)
TARGET = 'suicide_attempt'
X = df.drop(columns=[TARGET])
y = df[TARGET]
print(X.shape)
print(y.shape)
```

```python
# Training / test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2, stratify=y)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### First Attempt

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight='balanced'
)

rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix

ConfusionMatrixDisplay.from_estimator(rf, X_test, y_test)


cm = confusion_matrix(y_test, y_pred)

cm_df = pd.DataFrame(
    cm,
    index=["Actual 0", "Actual 1", "Actual 2"],
    columns=["Pred 0", "Pred 1", "Pred 2"]
)

print(cm_df)
```

```python
importances = pd.Series(rf.feature_importances_, index=X_train.columns)
print(importances.sort_values(ascending=False).head(10))
```
