```python
# For data manipulation
import pandas as pd
import numpy as np

# For plotting
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots appear inside the notebook
%matplotlib inline

df = pd.read_csv("../data/raw/Final_SP_dataSet.csv", keep_default_na=False)
```

# Data Cleaning

## Step 1: Convert column names to lowercase and snake case

```python
# See basic info: columns, data types, and null counts
print(df.info())

# Convert column names to lowercase and snake case
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print(df.info())
```

## Step 2: Verify columns are correct data types

```python
for col in df.columns:
    print(f"Column: {col}")
    print(df[col].apply(type).value_counts())
    print("-" * 40)
```

## Step 3: Check unique values in columns for missing/incorrect data

```python
for col in df.columns:
    print(df[col].value_counts())
```

```python
# Age of 221 is incorrect
# Changing it to the median age
median_age = df.loc[df['age'] < 100, 'age'].median()
df.loc[df['age'] == 221, 'age'] = median_age

# Found invalid answer 'pf' in depression_level column
# Changing it to 'Missing'
df.loc[df['depression_level'] == 'pf', 'depression_level'] = 'Missing'

# Clean strings
for col in df.columns:
    if col == 'age':
        continue
    else:
        df[col] = df[col].str.strip().str.title()

for col in df.columns:
    print(df[col].value_counts())
```

## Step 4: Encode categorical variables

### Ordinal Categories

```python
# Ordinal columns
ordinal_mappings = {
    'stress_level': {
        'Low': 0,
        'Moderate': 1,
        'High': 2
    },
    'academic_performance': {
        'Excellent': 0,
        'Good': 1,
        'Average': 2,
        'Poor': 3
    },
    'health_condition': {
        'Normal': 0,
        'Fair': 1,
        'Abnormal': 2
    },
    'depression_level': {
        'Sometimes': 0,
        'Often': 1,
        'Always': 2
    },
    'anxiety_level': {
        'Sometimes': 0,
        'Often': 1,
        'Always': 2
    },
    'self_harm_story': {
        'No': 0,
        'Yes': 1
    },
    'suicide_attempt': {
        'Never Thought': 0,
        'Thought': 1,
        'Attempted': 2
    }
}

# Apply ordinal mappings to the dataframe
for column, mapping in ordinal_mappings.items():
    df[column] = df[column].map(mapping)

# Set missing value to -1 and convert to int
df['depression_level'] = df['depression_level'].fillna(-1).astype(int)

for col in df.columns:
    print(df[col].value_counts())
```

### Nominal Categories

```python
nominal_columns = ['gender', 'relationship_condition', 'family_problem', 'mental_support']
# One-hot encoding of nominal columns; preserving all dummy columns to preserve interpretability
df = pd.get_dummies(df, columns=nominal_columns, drop_first=False)
print(df.columns)
```

```python
for col in df.columns:
    print(df[col].value_counts())
```

```python
df.isnull().sum().sort_values(ascending=False)
```

# Model Training

```python
# Set features (x) and target (y)
TARGET = 'suicide_attempt'
X = df.drop(columns=[TARGET])
y = df[TARGET]
print(X.shape)
print(y.shape)
```

```python
# Training / test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2, stratify=y)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### First Attempt

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight='balanced'
)

rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix

ConfusionMatrixDisplay.from_estimator(rf, X_test, y_test)


cm = confusion_matrix(y_test, y_pred)

cm_df = pd.DataFrame(
    cm,
    index=["Actual 0", "Actual 1", "Actual 2"],
    columns=["Pred 0", "Pred 1", "Pred 2"]
)

print(cm_df)
```

```python
importances = pd.Series(rf.feature_importances_, index=X_train.columns)
print(importances.sort_values(ascending=False).head(10))
```

### Second Attempt

```python
# Increase number of trees
rf2 = RandomForestClassifier(
    n_estimators=500,
    random_state=42,
    class_weight='balanced'
)

rf2.fit(X_train, y_train)

y_pred2 = rf2.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print('-' * 40)
print("Accuracy:", accuracy_score(y_test, y_pred2))
print("\nClassification Report:\n", classification_report(y_test, y_pred2))
```

### Third Attempt

```python
# Set max_depth to 10
rf3 = RandomForestClassifier(
    n_estimators=500,
    max_depth=10,
    random_state=42,
    class_weight='balanced'
)

rf3.fit(X_train, y_train)

y_pred3 = rf3.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print('-' * 40)
print("Accuracy:", accuracy_score(y_test, y_pred2))
print("\nClassification Report:\n", classification_report(y_test, y_pred2))
print('-' * 40)
print("Accuracy:", accuracy_score(y_test, y_pred3))
print("\nClassification Report:\n", classification_report(y_test, y_pred3))
```

# Charting

## Class distribution

```python
class_labels = {
    0: 'Never Thought',
    1: 'Thought',
    2: 'Attempted'
}

# Compute percent of total
percent_df = (
    df["suicide_attempt"]
    .value_counts(normalize=True)
    .sort_index()
    .mul(100)
    .reset_index()
)

percent_df.columns = ["suicide_attempt", "percent"]
percent_df["label"] = percent_df["suicide_attempt"].map(class_labels)

plt.figure(figsize=(6, 4))

ax = sns.barplot(
    data=percent_df,
    x="label",
    y="percent",
    order=[class_labels[0], class_labels[1], class_labels[2]]
)

ax.set_xlabel("Suicide Attempt Category")
ax.set_ylabel("Percent of Dataset")
ax.set_title("Distribution of Suicide Attempt Categories")

# Remove only top/right spines
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

# Explicit axis limits + ticks
y_max = percent_df["percent"].max()
upper_limit = int(y_max * 1.1)

ax.set_ylim(0, upper_limit)
ax.set_yticks(range(0, upper_limit + 1, 10))
plt.tick_params(axis='x', which='both', length=0)

# Add percent labels
for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%", padding=2)

# Export chart as PNG
# plt.savefig("suicide_attempt_distribution.png", dpi=300, bbox_inches="tight", facecolor="white")

plt.tight_layout()
plt.show()
```

## Feature importance

```python
importances = importances.sort_values(ascending=False)

top_importances = importances.head(10).sort_values()
plt.figure(figsize=(8, 5))

ax = top_importances.plot(kind="barh")

plt.title("Top 10 Feature Importances (Random Forest)")
plt.xlabel("Importance Score")
plt.ylabel("Feature")

# # Add spacing on right side
# max_val = top_importances.max()
# plt.xlim(0, max_val * 1.15)  # 15% extra space

# Remove y-axis ticks
plt.tick_params(axis='y', which='both', length=0)

# Remove only top/right spines
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

# Add value labels to each bar
for i, v in enumerate(top_importances):
    ax.text(
        v + 0.002,      # slight offset to the right of bar
        i,              # y-position
        f"{v:.3f}",     # format to 3 decimal places
        va="center"
    )

# Export chart as PNG
# plt.savefig("top_10_feature_importances.png", dpi=300, bbox_inches="tight", facecolor="white")

plt.tight_layout()
plt.show()
```

## Class distribution in stress_level column categories

```python
# Create cross-tab
stress_dist = pd.crosstab(
    df["stress_level"],
    df["suicide_attempt"],
    normalize="index"
) * 100

# Map class labels
stress_dist.columns = stress_dist.columns.map(class_labels)

# Map stress level labels
stress_label_map = {
    0: "Low",
    1: "Moderate",
    2: "High"
}
stress_dist.index = stress_dist.index.map(stress_label_map)

stress_levels = stress_dist.index.tolist()
classes = stress_dist.columns.tolist()

plt.figure(figsize=(8, 5))

bottom = np.zeros(len(stress_levels))

for cls in classes:
    values = stress_dist[cls].values
    plt.bar(
        stress_levels,
        values,
        bottom=bottom,
        label=cls,
        edgecolor="none"
    )
    bottom += values

plt.ylabel("Percent Within Stress Level")
plt.xlabel("Stress Level")
plt.title("Distribution of Suicide Attempt Categories by Stress Level")

ax = plt.gca()
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.tick_params(axis='x', length=0)

plt.legend(
    title="Suicide Attempt Category",
    bbox_to_anchor=(1.02, 1),
    loc="upper left",
    borderaxespad=0
)

# Export chart as PNG
# plt.savefig("distribution_by_stress_level.png", dpi=300, bbox_inches="tight", facecolor="white")

plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()
```
